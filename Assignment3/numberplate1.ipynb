```python
try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract
img = Image.open("temp1/car1.jpg") 

def ocr_core(img):
    text = pytesseract.image_to_string(Image.open(img))
    return text

print(ocr_core("temp1/car1.jpg"))
```
