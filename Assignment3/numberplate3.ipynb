```python
from PIL import Image 

im = Image.open('try3.jpg')

left = 332
top = 300
right = 560
bottom = 370

im1 = im.crop((left, top, right, bottom)) 

im1.save("temp1/car3.jpg")
Image.open("temp1/car3.jpg")
```

```python
try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract
img = Image.open("temp1/car3.jpg") 

def ocr_core(img):
    text = pytesseract.image_to_string(Image.open(img))
    return text

print(ocr_core("temp1/car3.jpg"))
```
